```python
from pathlib import Path
from datetime import datetime, time
import zoneinfo

import polars as pl
from matplotlib import pyplot as plt
from tqdm.auto import tqdm

from src.data.index import CandlesAssetData
```

```python
std_norms = []

for path in tqdm(list(Path("../data/candles_1_min/").glob("*.parquet"))):
    std_norm = (
        pl.read_parquet(path)
        .filter(pl.col("ts").dt.time() >= time(11, 0, 0))
        .filter(pl.col("ts").dt.time() <= time(18, 0, 0))
        .group_by(pl.col("ts").dt.date())
        .agg((pl.col("close").std() / pl.col("close").mean()).alias("std_norm"))
        .filter(pl.col("ts").dt.weekday() <= 5)
        # .plot.line(x="ts", y="std_norm")
        ["std_norm"].mean()
    )
    std_norms.append((path.stem, std_norm))
```
